```python
from flask import Flask, render_template, request, jsonify, Response, stream_with_context
import google.generativeai as genai

app = Flask(__name__)

# Configure API key (make sure to replace 'api_key' with your actual API key)
genai.configure(api_key='api here')

generation_config = {
    "temperature": 0.9,
    "top_p": 1,
    "top_k": 1,
    "max_output_tokens": 4096,
}

safety_settings = [
    {
        "category": "HARM_CATEGORY_HARASSMENT",
        "threshold": "BLOCK_MEDIUM_AND_ABOVE"
    },
    {
        "category": "HARM_CATEGORY_HATE_SPEECH",
        "threshold": "BLOCK_MEDIUM_AND_ABOVE"
    },
    {
        "category": "HARM_CATEGORY_SEXUALLY_EXPLICIT",
        "threshold": "BLOCK_MEDIUM_AND_ABOVE"
    },
    {
        "category": "HARM_CATEGORY_DANGEROUS_CONTENT",
        "threshold": "BLOCK_MEDIUM_AND_ABOVE"
    }
]

model = genai.GenerativeModel(model_name="gemini-1.0-pro",
                              generation_config=generation_config,
                              safety_settings=safety_settings)

@app.route("/", methods=["GET"])
def index():
    return render_template("index.html")

@app.route("/", methods=["POST"])
def generate_text_stream():
    input_data = request.get_json()
    prompt = input_data.get("prompt")

    if not prompt:
        return jsonify({"error": "Prompt is required"}), 400

    def generate_stream():
        if "hey siri" in prompt:
            response = model.generate_content(prompt + " generate it in hindi but only generate the anwer if it is mental health related"+"do it as if you are a therapist with years of experience  on mental health and speak sympathetically", stream=True)
            for chunk in response:
                yield chunk.text + "\n"
        else:
            response = model.generate_content("not applicable in hindi", stream=True)
            for chunk in response:
                yield chunk.text + "\n"

    return Response(stream_with_context(generate_stream()), mimetype="text/plain")

if __name__ == "__main__":
    app.run()
```
